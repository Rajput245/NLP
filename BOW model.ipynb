```python
dataset= nltk.sent_tokenize(sent)
```

```python
for i in range(len(dataset)):
    dataset[i]=dataset[i].lower()
    dataset[i]=re.sub(r'\s+'," ",dataset[i])
    dataset[i]=re.sub(r'\W'," ", dataset[i])
    dataset[i]=re.sub(r'\d'," ", dataset[i])
```

```python
dataset
```

# lets create a histogram

```python
wordcount={}
for data in dataset:
    words=nltk.word_tokenize(data)
    for word in words:
        if word not in wordcount.keys():
            wordcount[word]=1
        else:
            wordcount[word]+=1
```

```python
wordcount
```

```python
import heapq
```

```python
freq_word=heapq.nlargest(100,wordcount,key=wordcount.get)
```

```python
freq_word
```

```python
X=[]
for data in dataset:
    vector=[]
    for word in freq_word:
        if word in nltk.word_tokenize(data):
            vector.append(1)
        else:
            vector.append(0)
    X.append(vector)
```
